```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def crete_dataset(df, col_thr=100, row_thr=100):
    col_nan = pd.DataFrame(df.isnull().sum()*100/df.shape[0], columns=['percent_nan'])
    cols = col_nan[col_nan['percent_nan'] <= 100 - col_thr].index
    row_nan = df.isnull().sum(axis=1)*100/df.shape[1]
    rows = row_nan[row_nan <= 100 - row_thr].index
    return df.loc[rows, cols].reset_index(drop=True)
```

```python
df = pd.read_csv('df_outer_full.csv')
```

```python
df = crete_dataset(df, col_thr=100, row_thr=100)
```
